```python
# Named Entity Recognition. (NER)
```

```python
doc8=nlp("Apple to build a Hong  Kong Factory for $6 million")
```

```python
for token in doc8:
    print(token.text,end=" | ")
```

```python
print("\n----")
```

```python
for ent in doc8.ents:
    print(ent.text+"-"+ent.label_+"-"+str(spacy.explain(ent.label_)))
```
